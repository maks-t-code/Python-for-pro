### Задание 1: Анализ данных о студентах с использованием NumPy


**Описание:**\
У вас есть данные о студентах, включающие оценки по различным предметам. Используйте библиотеку NumPy для анализа этих данных.

```python
students_data = {
    'math': [85, 78, 92, 70, 88],
    'physics': [90, 82, 76, 85, 89],
    'informatics': [88, 92, 80, 87, 90]
}

students_data
```

**Задание:**
* Найдите среднюю оценку по каждому предмету.
* Найдите медианную оценку по каждому предмету.
* Вычислите стандартное отклонение по каждому предмету.
* Определите предмет с самой высокой средней оценкой.

```python
# Решение
import numpy as np

# Создадим массивы предметов и массивы оценок
subjects = []
for i in students_data.keys():
    subjects.append(i)
subjects_array = np.array(subjects)
grades = []
for i in students_data.values():
    grades.append(i)
grades_array = np.array(grades)

# Вычислим среднюю оценку по каждому предмету
average = np.mean(grades_array, axis=1)
print(f'Средняя оценка по математике: {average[0]}')
print(f'Средняя оценка по физике: {average[1]}')
print(f'Средняя оценка по информатике: {average[2]}')

# Вычислим медианную оценку по каждому предмету
median = np.median(grades_array, axis=1)
print(f'Медианная оценка по математике: {median[0]}')
print(f'Медианная оценка по физике: {median[1]}')
print(f'Медианная оценка по информатике: {median[2]}')

# Вычислим стандартное отклонение по каждому предмету
std_deviation = np.std(grades_array, axis=1)
print(f'Стандартное отклонение по математике: {std_deviation[0]}')
print(f'Стандартное отклонение по физике: {std_deviation[1]}')
print(f'Стандартное отклонение по информатике: {std_deviation[2]}')

# Определим предмет с самой высокой средней оценкой
most_grade_subject = subjects_array[np.argmax(average)]
print(f'Предмет с самой высокой средней оценкой: {most_grade_subject}')
```

### Задание 2: Анализ данных о продажах с использованием Pandas

**Описание:**\
Используйте библиотеку Pandas для анализа данных о продажах товаров. У вас есть данные о продажах товаров за последний год.

```python
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales': [15000, 17000, 16000, 18000, 14000, 19000, 22000, 21000, 23000, 25000, 24000, 26000]
}
sales_data
```

**Задание:**
* Найдите общий объем продаж за год.
* Найдите средний объем продаж в месяц.
* Найдите месяц с наибольшим и наименьшим объемом продаж.
* Постройте линейный график изменения объема продаж по месяцам (Matplotlib).

```python
# Решение

import pandas as pd
import matplotlib.pyplot as plt

# Преобразование массива данных в DataFrame
df_sales_data = pd.DataFrame(sales_data)

# Общий объем продаж за год
total_sales = df_sales_data['Sales'].sum()
print(f'Общий объем продаж за год: {total_sales}')

# Средний объем продаж в месяц
average_sales = df_sales_data['Sales'].mean()
print(f'Средний объем продаж в месяц: {average_sales}')

# Месяц с наибольшим и наименьшим объемом продаж
# Месяц с наибольшим объемом продаж
max_sales_month = df_sales_data.loc[df_sales_data['Sales'].idxmax(), 'Month']
print(f'Месяц с наибольшим объемом продаж: {max_sales_month}')
# Месяц с наименьшим объемом продаж
min_sales_month = df_sales_data.loc[df_sales_data['Sales'].idxmin(), 'Month']
print(f'Месяц с наименьшим объемом продаж: {min_sales_month}')

# График изменения объема продаж по месяцам
plt.figure(figsize=(11, 6))
plt.plot(df_sales_data['Month'], df_sales_data['Sales'])
plt.title('График изменения объема продаж по месяцам')
plt.xlabel('Месяцы')
plt.ylabel('Объем продаж')
```

### Задание 3: Задание на визуализацию данных с использованием Seaborn

**Описание:**\
Используйте готовый набор данных из библиотеки Seaborn для создания различных типов графиков. \
Ваша задача — загрузить датасет ***tips*** из готового набора Seaborn и построить несколько базовых графиков для анализа данных.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips') # используйте строго данный датасет

df.head()
```

Постройте следующие графики:
* Линейный график
* Столбчатый график
* Гистограмма
* Диаграмма рассеяния

Условия:
* Параметры визуализации на ваш выбор
* Графики должны нести смысловую нагрузку

```python
# Решение

# Линейный график зависимости времени прихода посетителей от дня
sns.set_style("darkgrid")
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='day', y='time')
plt.title('Зависимость времени прихода посетителей от дня')
plt.xlabel('День')
plt.ylabel('Время дня')
plt.show()

# Столбчатый график для общего счета в курящем и не курящем зале
plt.figure(figsize=(8, 6))
sns.barplot(data=df, y='total_bill', hue='smoker')
plt.title("Распределение общего счета по курящим и некурящим")
plt.ylabel('Общий счет')
plt.show()

# Построение гистограммы для размера чаевых с разделением по курящим и некурящим
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='tip', hue='smoker', multiple='stack', bins=10, kde=True)
plt.title("Распределение размера чаевых по курящим и некурящим")
plt.xlabel('Чаевые')
plt.ylabel('Частота')
plt.show()

# Диаграмма рассеяния чаевых от общего счета
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.title('Зависимость чаевых от общего счета')
plt.ylabel('Чаевые')
plt.xlabel('Общий счет')
plt.show()
```
